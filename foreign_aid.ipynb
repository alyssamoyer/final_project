```python
import pandas as pd
```

# Loading and Cleaning Data

```python
#read foriegn aid/life expectancy csv to pandas
aid_csv = "resources/life_aid_2000_2015.csv"
foriegn_aid_df = pd.read_csv(aid_csv)
foriegn_aid_df.head()
```

```python
#drop duplicate index column
foriegn_aid_df = foriegn_aid_df.drop(columns = ["index"]) 
```

```python
#read oil and gas csv to pandas
oil_csv = "resources/oil_gas.csv"
oil_df_full = pd.read_csv(oil_csv)
oil_df_full.head()
```

```python
#drop columns that will not be used
oil_df = oil_df_full[["year", "eiacty", "net_oil_exports", "net_oil_exports_value"]]

#filter for just the years that match the life expectancy dataframe
oil_df_00_13 = oil_df.loc[(oil_df['year'] > 1999) & (oil_df['year'] < 2014) ]

#lowercase and remove spaces/parenthesis from column names in order to best match the country names in the foriegn aid dataset
oil_df_00_13['eiacty'] = oil_df_00_13['eiacty'].replace(' ','', regex=True).str.lower().str.split('(').str[0]

oil_df_00_13.head()
```

```python
#merge (left) the foriegn aid/life expectancy dataframe with the oil dataframe on the columns country and year.
aid_oil = pd.merge(foriegn_aid_df, oil_df_00_13, how='left', left_on=['country', 'year'], right_on = ['eiacty', 'year'] )
aid_oil
```

```python
#show column counts
aid_oil.count()
```

```python
#filter the dataframe for just the year 2011
aid_oil_11 = aid_oil.loc[(aid_oil['year'] == 2011)]

aid_oil_11.head()
```
